## [House Prices - Advanced Regression Techniques](https://www.kaggle.com/c/house-prices-advanced-regression-techniques)
### A super popular data science problem avaliable on [kaggle](https://www.kaggle.com/). All most every biginner Data Science student has tryed to solve it.

### In this exaple usage tutorial we are going to compare  how regression algorithm is performing with no actual data cleaning and with de usage of the **maslow-proprocessig-toll**

```python
from package.classes import PreProcessor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
import pandas as pd
train = pd.read_csv("test/train.csv")
train.info()
```

### It consists in a dataframe with 79 columns representing independent variables, and 1 target variable called **Saleprice** (in this tutorial we call it *target column*). It has many missing values, and categorical.


## Method: remove_na_columns()
### The first problem we are going to deal with is the columns wich was a big proportion os missing data. Assuming this features arent valuable to our analysis we are deleting It.
### In the next lines of code we detemine that all columns with a higer percentage of missing values than 40% should be removed

```python
pre = PreProcessor(train)
pre.remove_na_columns(max_na_proportion=0.4) # Max proportion of missing values 40%
```

### As we can see the method remove_na_columns() droped 4 columns



## Method: replace_na_values()
### Now we want all the remainig columns that contains missing values to have it replced. So we call the method replace_na_values()

```python
pre.replace_na_values()
```

### The method replace_na_values() replaces all numeric columns with the median value of that columns, and replaces all categorical columns with the mde values of that column

## Method: encode_categorial()
### Now we use the method encode_categorical() to encode categorical columns

```python
pre.encode_categorical()
```

### Encode_categorical() uses pd.LabelEncoder to encode categorical columns 

## Method: normalize()
### To deal with skewness and kurtosis we use the function normalize, normalize method uses np.log or np.sqrt to normalize data reducing skewness and kurosis

```python
pre.normalize()
```

## Method feature_selction()
### Last but not less important, the method feature_selection() selects the colunmns that has a minimun absolute correlation with the target column, and drop the restant columns

```python
pre.feature_selection()
```

### As we can see, the melhod feature_selection selected 21 columns that presanted a correletion above 40% with the target column SalePrice.

## Now we have a clean and preprocessed dataframe to start playing with our the ML os NN tools...

```python
pre_processed_data = pre.df
pre_processed_data.head()
```

## Let's have som fun and use this celan and pre-processed data to make the cross validation os a SVR model

```python
from sklearn.model_selection import cross_validate
from sklearn.pipeline import Pipeline
scaler = StandardScaler()
pipe = Pipeline([('std_scaler', StandardScaler()), #Building a pipeline with a StandardScaler and a SVR regressor
                ('SVR', SVR())])


X , y = pre_processed_data.drop(columns=pre_processed_data.columns[-1]), pre_processed_data.iloc[:, -1] #Splitting our data betweeen independent and target variables
scores = cross_validate(pipe, X, y, cv=5, n_jobs=-1)

mean_score_accuracy = scores['test_score'].mean()
print(f"Model's mean accuracy: {round(mean_score_accuracy, 2)}%")

```
